```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('download.jpg')

if img is None:
    print("Error: Gambar 'tugas_mandiri.jpg' tidak ditemukan. Silakan upload dulu!")
else:
    # 1. Resize gambar menjadi 300x300
    img_300 = cv2.resize(img, (300, 300)) #

    plt.imshow(cv2.cvtColor(img_300, cv2.COLOR_BGR2RGB))
    plt.title('Operasi 1: Resize 300x300')
    plt.axis('off')
    plt.show()
```

```python
if 'img_300' in locals():
    # 2. Crop bagian tengah (mengambil koordinat 75 sampai 225)
    start, end = 75, 225
    img_cropped = img_300[start:end, start:end] #

    plt.imshow(cv2.cvtColor(img_cropped, cv2.COLOR_BGR2RGB))
    plt.title('Operasi 2: Crop Bagian Tengah')
    plt.axis('off')
    plt.show()
else:
    print("Jalankan Segmen 1 terlebih dahulu!")
```

```python
if 'img_cropped' in locals():
    # 3. Konversi ke Grayscale dan Tampilkan Histogram
    gray_cropped = cv2.cvtColor(img_cropped, cv2.COLOR_BGR2GRAY)

    hist_gray = cv2.calcHist([gray_cropped], [0], None, [256], [0, 256]) #

    plt.figure(figsize=(8, 5))
    plt.plot(hist_gray, color='black')
    plt.title('Operasi 3: Histogram Grayscale Bagian Tengah')
    plt.xlabel('Intensitas (0-255)')
    plt.ylabel('Jumlah Piksel')
    plt.show()
else:
    print("Jalankan Segmen 2 terlebih dahulu!")
```
